### Identify and analyze patterns in the data.

```python
import seaborn as sns
sns.pairplot(df)
```

### Visualize the distribution of key variables (e.g., Traffic_Index, Date).

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
sns.histplot(df['TrafficIndexLive'],kde=True,bins=20,color='blue')
plt.title('The distribution for Traffic_Index')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['TravelTimeHistoric'],kde=True,bins=20,color='blue')
plt.title('The distribution for TravelTimeHistoric')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['TrafficIndexWeekAgo'],kde=True,bins=20,color='blue')
plt.title('The distribution for TrafficIndexWeekAgo')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['TravelTimeLive'],kde=True,bins=20,color='blue')
plt.title('The distribution for TravelTimeLive')
plt.show()
```

### Explore relationships between variables (e.g., Traffic_Index vs. Weather_Condition).


```python
sns.boxplot(data=df,x="TrafficIndexLive",y="City")
```

```python
sns.boxplot(data=df,x="TravelTimeLive",y="City")
```
